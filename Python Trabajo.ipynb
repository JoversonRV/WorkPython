```python
#Cruzamiento de Variables (RegiÃ³n por OcupaciÃ³n)-AnÃ¡lisis de Chi-Cuadrado
#Tabla de contigencia:
tabla_contingencia1 = pd.crosstab(data['region'], data['P608'])

print(tabla_contingencia1)
```

```python
pd.crosstab(data['region'], data['P608'], normalize= "all")*100
```

```python
from scipy.stats import chi2_contingency

# Calcular chi-cuadrado, valor p, grados de libertad y frecuencias esperadas
chi2, p, dof, ex = chi2_contingency(tabla_contingencia1)

print(f"Chi-cuadrado: {chi2}")
print(f"p-valor: {p}")
print(f"Grados de libertad: {dof}")
print("Frecuencias esperadas:")
print(ex)
```
