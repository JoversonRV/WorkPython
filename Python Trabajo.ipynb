```python
#Generamos estadísticos descriptivos en base a las variables escogidas:
#Primero trabajamos con la variable categórica"P608", que nos muestra la ocupación principal o de negocio del encuestado:

# Frecuencias de p608 (Cargo o ocupación)
frecuencias_p608 = data['P608'].value_counts()
porcentajes_p608 = data['P608'].value_counts(normalize=True) * 100

print("Frecuencias de p608:\n", frecuencias_p608)
print("\nPorcentajes de p608:\n", porcentajes_p608)
```

```python
#Graficamos 

# Paso 3: Creamos las barras horizontales
plt.barh(y=porcentajes_p608.index, width=porcentajes_p608.values,
         color = 'skyblue') # Use .index to access the province names

# Detalles: Agregar título y etiquetas
plt.title('¿UD SE DESEMPEÑO EN SU OCUPACION PRINCIPAL O NEGOCIO COMO?(2010)')
plt.xlabel('Procentaje promedio')
plt.ylabel('Ocupación Principal o Negocio')
```

```python
#Hacemos lo mismo con la siguiente variable categórica "P619", que indica la actividad que realizó el encuestado para conseguir trabajo la semana pasada
frecuencias_p619 = data['P619'].value_counts()
porcentajes_p619 = data['P619'].value_counts(normalize=True) * 100

print("Frecuencias de p619:\n", frecuencias_p619)
print("\nPorcentajes de p619:\n", porcentajes_p619)
```

```python
# Paso 3: Creamos las barras horizontales
plt.barh(y=porcentajes_p619.index, width=porcentajes_p619.values,
         color = 'green') # Use .index to access the province names

# Detalles: Agregar título y etiquetas
plt.title('¿QUÉ HIZO LA SEMANA PASADA PARA CONSEGUIR TRABAJO? (2010)')
plt.xlabel('Procentaje promedio')
plt.ylabel('Instancia o persona a la que acudió')
```

```python
#Una variable categórica adicional que nos servirá para el cruzamiento es la región de la que proviene el encuestado "region"
frecuencias_region = data['region'].value_counts()
porcentajes_region = data['region'].value_counts(normalize=True) * 100

print("Frecuencias de region:\n", frecuencias_region)
print("\nPorcentajes de region:\n", porcentajes_region)
```

```python
#Graficamos la tercera variable categórica
# Paso 3: Creamos las barras horizontales
plt.barh(y=porcentajes_region.index, width=porcentajes_region.values,
         color = 'green') # Use .index to access the province names

# Detalles: Agregar título y etiquetas
plt.title('REGION DEL ENCUESTADO (2010)')
plt.xlabel('Procentaje promedio')
plt.ylabel('Región')
```

```python
#Cruzamiento de Variables (Región por Ocupación)-Análisis de Chi-Cuadrado
#Tabla de contigencia:
tabla_contingencia1 = pd.crosstab(data['region'], data['P608'])

print(tabla_contingencia1)
```

```python
pd.crosstab(data['region'], data['P608'], normalize= "all")*100
```

```python
from scipy.stats import chi2_contingency

# Calcular chi-cuadrado, valor p, grados de libertad y frecuencias esperadas
chi2, p, dof, ex = chi2_contingency(tabla_contingencia1)

print(f"Chi-cuadrado: {chi2}")
print(f"p-valor: {p}")
print(f"Grados de libertad: {dof}")
print("Frecuencias esperadas:")
print(ex)
```
